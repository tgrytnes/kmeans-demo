# ðŸ§  K-Means Clustering on Tips Dataset

This notebook demonstrates how K-Means clustering can help group restaurant customers based on their total bill and tip amount.

We use a real dataset and display the clusters visually.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)
df.head()
```

```python
X = df[['total_bill', 'tip']]

kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
colors = ['red', 'blue', 'green']
for c in sorted(df['Cluster'].unique()):
    cluster_data = df[df['Cluster'] == c]
    plt.scatter(cluster_data['total_bill'], cluster_data['tip'], label=f'Cluster {c}', color=colors[c])
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.title('Customer Segments (K-Means Clustering)')
plt.legend()
plt.grid(True)
plt.show()
```

## ðŸ“Š Interpretation

The K-Means clustering algorithm has grouped the restaurant customers into three segments based on their total bill and tip amount:

- **Cluster 0 (Red):** Medium spenders with medium tips. Likely the "typical" customer.
- **Cluster 1 (Blue):** Low spenders with low tips. Possibly budget-conscious diners.
- **Cluster 2 (Green):** High spenders with generous tips. This group might represent premium or group customers.

This analysis demonstrates how unsupervised learning can uncover hidden patterns in simple business data.
